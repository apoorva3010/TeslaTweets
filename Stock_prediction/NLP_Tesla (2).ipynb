```
dataset = pd.read_csv("dataset_with_target.csv",encoding='windows-1254') 
```

```
import re
import nltk
nltk.download('stopwords')
```

```
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```
dataset.shape
```

```
dataset.info()
```

```
new_tweet = []
for i in range(0, 633):
  tweet = re.sub('[^a-zA-Z]', ' ', dataset['Tweet'][i])
  tweet = tweet.lower()
  tweet = tweet.split()
  ps = PorterStemmer()
  tweet = [ps.stem(word) for word in tweet if not word in set(stopwords.words('english'))]
  tweet = ' '.join(tweet)
  new_tweet.append(tweet)
```

```
print(new_tweet)
```

```
len(new_tweet)
```

```
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1000)
X = cv.fit_transform(new_tweet).toarray()
y = dataset.iloc[:, -1].values
```

```
X.shape
```

```
df = pd.DataFrame(X) 
```

```
df.to_csv('newdata_with_target.csv')
```

```
pd.read_csv('newdata_with_target.csv')
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10, random_state = 0)
```

```
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
```

```
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
model_one = LogisticRegression()
```

```
model_one.fit(X_train,y_train)
```

```
y_pred_one = model_one.predict(X_test)
```

```
print(np.concatenate((y_pred_one.reshape(len(y_pred_one),1), y_test.reshape(len(y_test),1)),1))
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_one)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_one)
```
